```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
def get_spacy_vec(text):
    doc = nlp(text)
    return doc.vector/doc.vector_norm
```

```python
vec_1 = get_spacy_vec('thunderstorm with light drizzle')
vec_2 = get_spacy_vec('thunderstorm with light rain')
vec_3 = get_spacy_vec('sky is clear')
```

```python
np.dot(vec_1, vec_2)
```

```python
np.dot(vec_1, vec_3)
```
